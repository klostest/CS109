# Historical temperature data and degree day sums

For centuries, plant phenology has been recognized to depend on degree day temperature sums, and this is still a widely used model for phenology forecasts and hindcasts (Richardson et al, 2006).  To train the model on historical phenology data, the DAYMET data set of spatially explicit climatic data will be used: http://daymet.ornl.gov/.  DAYMET data including a region approximately 100km by 100km, centered on Boston, was obtained to demonstrate the concept outlined in this project.  DAYMET data is available in NetCDF format (http://en.wikipedia.org/wiki/NetCDF), however we have reformatted the data as numpy variables for this project.  The following cell shows how the DAYMET data is used to calculate spatially explicit degree day sums.  There are two DAYMET arrays for daily minimum and maximum temperature at each DAYMET grid point, averaged to approximate mean daily temperature.  Each grid point corresponds to a geographical location with associated latitude and longitude.  This data is used to calculate the degree day sum (GDD) for day of year (DOY) t, which represents accumulated warmth above a threshold temperature $T_{base}$ over the course of the year:

$$GDD(t) = \sum_{t=1}^{365} f(t) + f(t-1)$$

where

$$f(t)= \mathrm{min} \left( \frac{ T_\mathrm{max}(t) + T_\mathrm{min}(t) }{2}-T_\mathrm{base} , 0 \right)$$

```python
def DAYMET_wdd(year, baseTemp, start_doy, chill):
    #Uses the DAYMET data to compute an array of degree days, with start date start_doy
    #This function relies on the Nio api, which requires installation.  For grading purposes
    #we have linked the output files from this function that can be downloaded to avoid the
    #need for this.  For more info on Nio, see http://www.pyngl.ucar.edu/Nio.shtml
    #
    #inputs
    #year (int), the year to calculate warming degree days
    #baseTemp (int), the base temperature for the degree day sum
    #start_doy (int), day of year to begin the summation on
    #chill (str), 'yes' or 'no' to indicate whether using model in chilling formulation.
    #
    #output
    #wdd (numpy array), a (365, 252, 216) array with the degree day sum for each
    #day of the year at each DAYMET grid point
    
    #open the netcdf files with DAYMET data
    tmax_file = Nio.open_file('./DAYMET/' + str(year) +'/tmax.nc')
    tmin_file = Nio.open_file('./DAYMET/' + str(year) +'/tmin.nc')
    
    #get the temperature and coordinates variables as NioVariables
    tmax_var = tmax_file.variables['tmax']
    tmin_var = tmin_file.variables['tmin']
    
    lat_var1 = tmax_file.variables['lat']
    lon_var1 = tmax_file.variables['lon']
    
    #put values from NioVariables in numpy arrays
    tmax = tmax_var.get_value()
    tmin = tmin_var.get_value()
    tavg = (tmax + tmin)/2
    
    daymet_lat = lat_var1.get_value()
    daymet_lon = lon_var1.get_value()
    
    DAYMET_data_shape = np.shape(tavg)
    # (365, 252, 216)
    
    #Initialize
    wdd = np.zeros(DAYMET_data_shape)
    
    startDateIndex = start_doy - 1;
    
    for i in range(startDateIndex, DAYMET_data_shape[0]):
        #How much above the base temp is the avg temp
        thisDay = tavg[i,:,:] - baseTemp
        if chill == 'yes':
            #Express chilling below the baseTemp as a positive sum
            thisDay = -thisDay
            
        #Turn negatives to zero
        thisDay[thisDay<0] = 0
        if i == 0:
            #First day of year
            wdd[i,:,:] = thisDay
        else:
            #Add to previous day's sum
            wdd[i,:,:] = thisDay + wdd[i-1,:,:]
    
    return wdd
```

# Phenology data

We obtained phenology data for the region around Boston from the MODIS (moderate resolution imaging spectroradiometer) instrument.  Two MODIS radiometers are mounted on satellites and measure light reflected from the earth's surface, at 500 m spatial resolution, in seven wavebands, including visible light in the red, green, and blue regions, as well as infrared light.  The data is available via the MODIS web service (https://daac.ornl.gov/MODIS/MODIS-menu/modis_webservice.html) and was processed in Matlab using software the group had developed for a previous project:  https://github.com/klostest/PhenoCamAnalysis/tree/master/phenoFit_v1_0 (Klosterman et al, unpublished).  

Roughly speaking, we created MODIS time series representing seasonal vegetation activity, using visible and infrared surface reflectances with 8 day temporal resolution.  The data was smoothed and phenological dates were estimated by a threshold crossing method.  From these tresholds, we identify five key phenology dates:  the beginning (bud break) and end of leaf expansion in spring, as well as the beginning of leaf coloring, the most intense leaf coloring, and the end of leaf drop in autumn.  The process for date estimation was similar to that described here:  http://www2.geog.ucl.ac.uk/~plewis/geogg124/modelling.html#models-of-phenology.

To show the extent of the MODIS data, here are images of the red, green and blue bands on a few representative dates before, during, and after leaf expansion, and during autumn in 2004:

Since we will use degree day sums to predict phenology, and get the phenology data from MODIS, we must first harmonize these two spatially explicit data sets.  The MODIS data are associated with 500 meter pixels, which are land areas which look like a square, 500 meters on a side, in the sinusoidal projection used by MODIS:  https://lpdaac.usgs.gov/products/modis_overview .  In more typical map projections, these areas look like parallelograms.

We harmonize the MODIS and DAYMET data by taking the geographic coordinates of the center of each of the 40401 MODIS pixels in the study area, and finding the indices of the DAYMET grid point closest to it.

```python
#For each MODIS pixel, get the distance to the nearest DAYMET point, and the\
#corresponding indices
nPixels = 40401
minIndices = np.empty([nPixels,2])
minDist = np.empty(nPixels,)

#Load MODIS and DAYMET coordinates from file
MODIS_coords = np.genfromtxt('./MODIS/CS_109_MODIS_Lat_Lon.csv', delimiter=',')
daymet_lat = np.load('./DAYMET/daymet_lat.npy')
daymet_lon = np.load('./DAYMET/daymet_lon.npy')

for i in range(0,nPixels):
#     print coord
    #for each MODIS pixel, calculate the distance from all DAYMET pixels
    distFromDaymet = np.sqrt( np.power( (MODIS_coords[i,0] - daymet_lat), 2) \
                             + np.power( (MODIS_coords[i,1] - daymet_lon), 2 ) )
#     print np.shape(distFromDaymet)
    #find the value and index of minimum distance
    minDist[i] = np.min(distFromDaymet)
    
    #Create boolean array to find indices of this grid point
    minDistLogical = distFromDaymet == minDist[i]
    zerosToFindMin = np.zeros(np.shape(minDistLogical))
    zerosToFindMin[minDistLogical] = 1
    minIndices[i,:] = np.nonzero(zerosToFindMin)
```

# Combining the temperature and phenology data

Now that we have the geographic correspondence between the MODIS and DAYMET data, here is a function that returns the degree day sum for a given day of year, at every MODIS pixel

```python
#Function to get degree day sums for given day of year
def GDD_for_DOY(wdd, minIndices, DOY, nPixels):
    #Inputs
    #wdd, (365,n,n) numpy array, is the DAYMET derived warming degree days
    #minIndices, (nPixels,2) array, is correspondence between DAYMET and MODIS grid points
    #DOY is the desired DOY to get GDDs
    #nPixels is the number of MODIS pixels
    #Outputs
    #GDD_for_DOY is (nPixels,) with the GDDs for each pixel on the given DOY
    
    GDD_for_DOY = np.empty([nPixels,])
    
    for i in range(0,nPixels):
        #the spring date for this MODIS pixel
#         thisDate = springDates[i]
        #Subtract one for python indexing
        GDD_for_DOY[i] = wdd[DOY-1, minIndices[i,0], minIndices[i,1]]
    return GDD_for_DOY
```

Here is another function for looking up the number of degree days which have been accumulated by a certain phenological transition.

```python
def getGDDs_byThresh(transThresh, wdd, MODIS_dates, minIndices):
    #For each MODIS pixel, get the growing degree days for a given phenological transition,
    #using data from the the closest DAYMET grid point
    #
    #inputs
    #transThresh (int), between 1 and 9, corresponding to the 0.1, 0.2, ... 0.9 thresholds
    #wdd (numpy array), output from DAYMET_wdd above
    #MODIS_dates (numpy array), loaded from CSV above.
    #minIndices (numpy array), indices for geographical correspondence of MODIS and DAYMET
    #grid points
    
    #Need to subtract one to index the column
    dates = MODIS_dates[:,transThresh-1]
    #Round down partial days
    dates = np.floor(dates)
    
    #What is the number of MODIS pixels?
    nPixels = np.shape(MODIS_dates)[0]
    #initialize variable
    thisGDD = np.empty(nPixels,)
    
    #Spring
    #For each MODIS pixel
    for i in range(0,nPixels):
        #the spring date for this MODIS pixel
        thisDate = dates[i]
        thisGDD[i] = wdd[thisDate-1, minIndices[i,0], minIndices[i,1]]
    return thisGDD
```

# Preparing to fit the model

We choose a logistic regression to describe leaf phenology.  This model is analagous to the voting prediction in homework two.  In this case, a section of land surface has the opportunity to "vote" each day on whether to advance to the next phenological stage, based on the degree day sum.  To implement this we need time series with the current phenological stage for each MODIS pixel:

```python
def get_phenoStatus(MODIS_dates, doy_vec):
    #Create a vector for each DOY with the phenological status of every
    #MODIS pixel
    #0 for not having undergone the 0.1 transition, 1 after, then 2, 3, etc up to 9
    #
    #Inputs
    #
    #MODIS_dates, array with dates of transitions for every MODIS pixel
    #
    #doy_vec, days of year to make the phenoStatus array
    #
    #Outputs
    #
    #phenoStatus, nPixels x 200 array with the status on these DOYs
    
    nRows = np.shape(MODIS_dates)[0]
    phenoStatus = np.empty([nRows, len(doy_vec)])
    
    #Round down partial days to use as indices
    MODIS_dates_floor = np.floor(MODIS_dates)
    #Print example row of dates
#     print MODIS_dates_floor[0,:]
    
    #To use MODIS dates as indices corresponding to input DOY vector, necessary to
    #subtract the start date of the doy series minus one
    MODIS_dates_floor = MODIS_dates_floor - (min(doy_vec) - 1);

    for i in range(0, nRows):
        #There are 9 dates but this could be generalized
        for j in range(0, 9):
#             print j
            if j == 0:
                phenoStatus[i, 0:(MODIS_dates_floor[i, j]-2)] = j
            elif j < 8:
                phenoStatus[i, (MODIS_dates_floor[i, j-1]-1):(MODIS_dates_floor[i, j]-1)] = j
            elif j == 8:
                phenoStatus[i, (MODIS_dates_floor[i, j-1]-1):(MODIS_dates_floor[i, j]-1)] = j
                phenoStatus[i, (MODIS_dates_floor[i, j]-1):200] = 9

    return phenoStatus
```

Now we have the data which we hope to predict: phenological progressions through time.  This data will be reformulated into binary data to train the logistic regression.  At this point we need the features to train the logistic, which are degree day sums for all the MODIS pixels.  In this function, we also create a quality control mask.  MODIS measures the entire surface of the earth, including areas where we would not expect to observe deciduous tree phenology, such as the ocean and developed urban spaces.  In these areas the MODIS date estimation algorithm tends to fail to identify transition dates.  Since zeros are also used as fill values in the MODIS and DAYMET data, we make a combined zero mask that will be used to make sure we are only training the model using data from locations where there are probably deciduous trees, and we have temperature data.

```python
def make_degree_day_feature(wdd, minIndices, MODIS_dates, nPixels, doy_vec):
    #Make array of degree day sums for each MODIS pixel for spring time, and a mask
    #for later exclusion of fill values
    #
    #Inputs
    #
    #wdd, 3d numpy array with warming degree day sums for each day of year, for each
    #DAYMET grid point
    #
    #minIndices, (nPixels,2) array with indices of the DAYMET grid point corresponding
    #to each MODIS grid point
    #
    #MODIS_dates, numpy array input from csv
    #
    #nPixels, number of MODIS pixels
    #
    #doy_vec, numpy array of DOYs
    #
    #Outputs
    #
    #spring_GDDs (nPixels, len(doy_vec)) array of growing degree day sums
    #
    #zeroMask (nPixels,), logical array with True where both the MODIS date algorithm
    #successfully extracted a date, i.e. does not have the fill value of zero
    #, and the DAYMET degree day sum is not zero, and False elsewhere
    
    #Mask all MODIS dates for a pixel if any of them weren't
    #successfully estimated
    #This assumes three thresholds but could be generalized
    #Make a MODIS mask based on springDates for all pixels that have a fill value for any transition
    for i in range(0,9):
        if i == 0:
            MODIS_zero_mask = (MODIS_dates[:,i] != 0)
        else:
            MODIS_zero_mask = MODIS_zero_mask & (MODIS_dates[:,i] != 0)
    
    #Compute feature
    #on unmasked springDates array to preserve MODIS index
    #correspondence with DAYMET data
    GDDs = np.empty([nPixels,len(doy_vec)])
    
    #Make a similar DAYMET zero mask
    for i in range(0,len(doy_vec)):
        GDDs[:,i] = GDD_for_DOY(wdd, minIndices, doy_vec[i], nPixels)
        if i == len(doy_vec)-1:   #Apparently fill values, use on last day of analysis
            DAYMET_zero_mask = (GDDs[:,i] != 0)
    
    #Combine MODIS and DAYMET masks
    zeroMask = MODIS_zero_mask & DAYMET_zero_mask
    
    return GDDs, zeroMask
```

We are ready now to define the function that will create the logistic regressions and train them.  The trained models will be exported for use in phenology predictions on the web site.

```python
def logRegs(feature, binary, reg=0.0001):
    #Description:  this function takes a degree day sum feature and a transition time series
    #output from the phenoStatusSpring function
    #It then trains a separate logistic regression model using a binary for each transition
    #
    #Inputs
    #
    #feature (nPixels, t) array of growing degree day sums as described above
    #
    #binary, an (nPixels, t) array where t is the number of tresholds
    #input in transThreshes.  Has a one or a zero as described above
    #
    #**** I think binary is actually an array with [0 0 0 1 1 2 ... 9]
    #which here is turned into the array of ones and zeros
    #
    #reg (float), regularization parameter as described above, default=0.0001
    #
    #Outputs
    #
    #logReg, list (t items) of logistic regression objects trained on the input
    #feature and binary

    #Empty list
    logReg = []
    #How many models to fit? (how many classes, or phenological transitions, are there?)
    nModels = np.int(np.max(binary))
    print nModels
    
    for i in range(0, nModels):
        #append list with a new regression object
        temp = LogisticRegression(C=reg)
        logReg.append(temp)
        
        #Make a binary classification for each transition.  Ones if the time series is
        #past the transition, zeros if not.
        this_binary = np.copy(binary)
        this_binary[this_binary<i+1] = 0
        this_binary[this_binary>=i+1] = 1
        
        #This should be just zeros and ones now
        print np.unique(this_binary)
        #Fit the model
        logReg[i].fit(feature, this_binary)

    return logReg
```

To illustrate how the model works, we train the spring model on a single year of data.  A similar approach is taken for autumn phenology, except using chilling degrees days.  

```python
#We have already been using the warming degree day sum for data exploration.  Here are the
#parameters that generated it, taken from Richardson et al 2006.
#Warming degree day sum for each DAYMET point, for each DOY
    # "from January 1 (spring)"
    # "HDD4 provides the best fit"
#use warming, not chilling degree days for spring

#Get chilling degree days for autumn
# "Heating degree-day sums (HDDk) are based on degree-days above the base
# temperature, k, whereas chilling degree-days (CDDk) are based on degree-days
# below the base temperature, from January 1 (spring) and August 15 (autumn)."
# "For autumn phenology, Model 2 works reasonably well with any base temperature ≥20°C"
# August 15 ~ DOY 227
# baseTemp = 20; start_doy = 227; chill = 'yes';
# cdd_07 = DAYMET_wdd(year, baseTemp, start_doy, chill)
# np.save('./DAYMET/cdd_07', cdd_07)

cdd_07 = np.load('./DAYMET/cdd_07.npy')
```

```python
#Apply zero masks
maskedFeature_spring = spring_GDDs[spring_zeroMask,:]
maskedBinary_spring = phenoStatus_spring[spring_zeroMask,:]

maskedFeature_fall = fall_GDDs[fall_zeroMask,:]
maskedBinary_fall = phenoStatus_fall[fall_zeroMask,:]

#Reshape for modeling
maskedFeature_spring = np.reshape( maskedFeature_spring, \
                                  [ np.shape(maskedFeature_spring)[0]*np.shape(maskedFeature_spring)[1] , 1 ] )
maskedBinary_spring = np.reshape( maskedBinary_spring, \
                                 [ np.shape(maskedBinary_spring)[0]*np.shape(maskedBinary_spring)[1] , ] )

maskedFeature_fall = np.reshape( maskedFeature_fall, \
                                  [ np.shape(maskedFeature_fall)[0]*np.shape(maskedFeature_fall)[1] , 1 ] )
maskedBinary_fall = np.reshape( maskedBinary_fall, \
                                 [ np.shape(maskedBinary_fall)[0]*np.shape(maskedBinary_fall)[1] , ] )

# print np.shape(maskedFeature), np.shape(maskedBinary)
```

```python
#Fit the model for spring
logReg_spring = logRegs(maskedFeature_spring, maskedBinary_spring, reg=0.0001)
# print np.unique(maskedBinary)
```

```python
#Fit the model for fall
logReg_fall = logRegs(maskedFeature_fall, maskedBinary_fall, reg=0.0001)
# print np.unique(maskedBinary)
```

```python
"""
Function
--------
get_lat_long_for_zip

Returns a tuple of the latitude and longitude. 

Inputs
-------
zip_code : string
    The ZIP code is a string since it can have leading zeros.
    
Examples
--------
>>> get_lat_long_for_zip('02138')
>>> (42.377045, -71.125611)
"""

def get_lat_long_for_zip(zip_code):
    
    if DEBUG: print "**EXEC: get_lat_long_for_zip(zip_code):", zip_code
    
    url = "http://api.geonames.org/findNearbyPostalCodesJSON?username=%s&country=US&postalcode=%s" % (geoname_api_username, zip_code)
    r = requests.get(url)
    data = r.json()
    if('status' in data and data['status']['value'] == 17):
        raise RuntimeError('Invalid ZIP code')
    lat = data["postalCodes"][0]['lat']
    lng = data["postalCodes"][0]['lng']
    return (float(lat), float(lng))


get_lat_long_for_zip('02238')
#get_lat_long_for_zip('02138')
```

# Utility caching methods

The weather apis in general have to be used carefully during development process. They hava a strict quota on the calls that you can make every day as well as restricitons on how many times you can call per second. To alleviate these restrictions (and as a valid optimization technique), we will cache the data so it is fetched only on the first call. On all subsequent calls, we will use the local cached data.

```python
"""
Setup utility methods to read and write data to cache.
The calling methods will take care of file nameing conventions.
"""

def read_cache_data_if_any(cache_file):
    if USE_LOCAL_CACHE == False:
        return None
    
    #Please create a cache dir if you want to cache in a different dir
    #cache_file = 'cache/' + cache_file; 
    
    if not os.path.isfile(cache_file):
        return None
    
    json_file_data = open(cache_file)
    json_data = json.load(json_file_data)
    if DEBUG: print "** reading from local cache", cache_file
    return json_data
    

def write_cache_data(json_data, cache_file):
    
    #cache_file = 'cache/' + cache_file; 
    
    try:
        with open(cache_file, 'w') as f:
            json.dump(json_data, f, ensure_ascii=False)
        return True
    except:
        return False
```

#Degree day sum calculation
As discussed above, a degree day is a measure of heating or cooling. Growing degree days (GDD) are calculated by taking the average of the daily maximum and minimum temperatures compared to a base temperature, Tbase, (in our case 4°C for heating dd and 20°C for chilling dd). Every day the following mathematical expression is added to a running sum of degree days for that year:

$$GDD = \frac{T_\mathrm{max}+T_\mathrm{min}}{2}-T_\mathrm{base}.$$

Similarly we can also calculate chilling days to find out heat accumulation for the fall events

$$Chilling DD = T_\mathrm{base} - \frac{T_\mathrm{max}+T_\mathrm{min}}{2}.$$

- To get the heating degree days sum we get the high and low temperature from the begining of the year  - for example January 1, 2013 if we are trying to get predictions for 2013. 

- To get the chilling degree days sum we get the high and low temperature from the August 15 of the year  - for example August 15, 2013 if we are trying to get predictions for 2013. 

Following is a listing of the methods used for degree day sum calculation:

####- get_weather_forecast(zip_code)

 - Input
   - zip_code: the zip code for which we want the weather forecast

 - Output
   - returns an array of tuples with daily minimum and maximum temperatures for next 10 days

####- get_weather_historic(zip_code, start_date, end_date)

  - Input
    - zip_code: the zip code for which we want the weather forecast
    - start_date is a date, the day to start accumulating degree days
    - end_date is mortly todays date. If not passed in, todays date is used

  - Output
    - returns an array of tuples with daily minimum and maximum temperatures for the start end range

####- degree_day_sum(zip_code, start_date, base_temp)
This is the main method which returns the forecasted degree day sums for the desired zip code. Degree day sum  calls the forecast and history methods to get the max/min temperatures and uses them to find the degree sum days. There are some other complications involved in getting historical weather data which we will describe in a later section. 

  - Input
    - start_date is a date, the day to start accumulating degree days
    - base_temp is an int, the base temperature

  - Output
    - deg_day_sums is a 1 by 10 array with forecasted degree day sums for the next 10 days (including today)


```python

"""
Function
--------
get_degree_day_sum

Returns the degree day sums for the historic weather condition leading upto today.

Inputs
-------
zip_code : string
    The zip code for the weather data
start_date : string
    Date in the format Year-Month-Date ('2013-1-14')
base_temp : float
    Base temperature used in the formula to calculate degree day sum.
chilling : boolean
    Retrieve chilling days using the chilling aggregation formula
    
Returns
-------
degree_sums : an array of tuples
  An array of degree days sums 
     
     
Examples
--------
>>> get_degree_day_sum('94538', '2013-10-1', 50)
>>> [582.5, 585.0, 589.5, 594.0, 600.0, 605.5, 607.5, 611.5, 616.0, 622.0]
"""

import datetime

# Helper method to calculate the degree day
def get_degree_day(temp_tuple, base_temp=5, chilling=False):
    
    if chilling:
        return get_chilling_day(temp_tuple)

    if DEBUG: print "**EXEC: get_degree_day(temp_tuple, base_temp=5)", temp_tuple, base_temp
    #if DEBUG: print "temp_tuple=", temp_tuple
        
    high_temp = float(temp_tuple[1])
    low_temp = float(temp_tuple[2])
    base_temp = float(base_temp)
    
    
    degree_day = (high_temp + low_temp)/2 - base_temp
    
    if degree_day < 0:
        degree_day = 0 

    if DEBUG: print "high_temp, low_temp, base_temp, degree_day", high_temp, low_temp, base_temp, degree_day
        
    return degree_day


# Helper method to calculate the degree day
def get_chilling_day(temp_tuple, base_temp=20):

    if DEBUG: print "**EXEC: get_degree_day(temp_tuple, base_temp=20)", temp_tuple, base_temp

    high_temp = float(temp_tuple[1])
    low_temp = float(temp_tuple[2])
    base_temp = float(base_temp)
    
    chilling_day = base_temp - (high_temp + low_temp)/2
    
    #if chilling_day < 0:
    #    chilling_day = 0 

    if DEBUG: print "high_temp, low_temp, base_temp, chilling_day", high_temp, low_temp, base_temp, chilling_day
        
    return chilling_day



# using world weather online site (wonline)
def get_degree_day_sum_wonline(zip_code, start_date, base_temp=5, end_date=None, chilling=False):
        
    if DEBUG: print "**EXEC: get_degree_day_sum_wonline(zip_code, start_date, base_temp=5, end_date=None, chilling=False)", \
        zip_code, start_date, base_temp, end_date, chilling
    
    hist_dict = {}
    hist_dates = create_monthly_history_dates(start_date, end_date)

    if DEBUG: print "history query dates:", hist_dates
    
    count = 0
    for hist_date in hist_dates:
        h_start_date = hist_date[0]
        h_end_date = hist_date[1]
        
        if DEBUG: print "** getting data for", h_start_date, "-", h_end_date
        
        hist_arr = get_weather_historic_wonline(zip_code, h_start_date, h_end_date)
        
        if DEBUG: print "\n** historic:", hist_arr
        
        hist_dict[count] = hist_arr
        count += 1
    
    #get historical degree day sum 
    degree_days = []
    yesterday_historic_temp = None
    for k, v in hist_dict.iteritems():
        for daily_temp_h in v:
            yesterday_historic_temp = daily_temp_h
            degree_day = get_degree_day(daily_temp_h, base_temp, chilling)
            if DEBUG: print "daily_temp=", daily_temp_h, "degree_day=", degree_day
            degree_days.append(degree_day)

    if len(degree_days) > 0:
        degree_days = degree_days[:-1] # remove last day - it is added to the forecast
    
    hist_degree_days_sum = sum(degree_days)
    if DEBUG: print "\n**historical degree days sum:", hist_degree_days_sum, "\n"

    # create array for forecast degree days sum
    forecast_degree_day_sum = []
    
    #append yesterday's data to it
    #forecast_degree_day_sum.append(last_historic_degree_day)
    
    #lookup forecast for next 10 days
    forecast_arr = get_weather_forecast_wonline(zip_code)
    
    #if we are not forecasting for the future (next 10 days)
    #we are probably just trying to look at the forecast for past 
    # i.e. forecast past (fp) - kinda wonky name
    if end_date is not None:
        end_date_date = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()
        date_10_days_ago =  datetime.date.today() - datetime.timedelta(days=10)

        if end_date_date <  date_10_days_ago: 
            fp_dates = create_historic_forecast_dates(end_date)
            
            forecast_arr = [] #reinitialize
            for fp in fp_dates:
                fp_start_date = fp[0]
                fp_end_date = fp[1]
                fp_results = get_weather_historic_wonline(zip_code, fp_start_date, fp_end_date)
                for fp_result in fp_results: 
                    forecast_arr.append(fp_result)
                    
    # now tag along yesterdays (the first day before fp_dates)
    # the reason is that it is used to break up cumulative probablitlies into absolutes
    if yesterday_historic_temp is not None:
        forecast_arr.insert(0, yesterday_historic_temp)
 
    if DEBUG: print "\n** forecast:", forecast_arr
    rolling_degree_day_sum = hist_degree_days_sum
    for daily_temp_f in forecast_arr:
        degree_day = get_degree_day(daily_temp_f, base_temp, chilling)
        rolling_degree_day_sum += degree_day
        degree_day_dict = {} #dict
        degree_day_dict[daily_temp_f[0]] = rolling_degree_day_sum #dict
        forecast_degree_day_sum.append(degree_day_dict) #dict
        #forecast_degree_day_sum.append(rolling_degree_day_sum)
        if DEBUG: print "degree day:", degree_day, "rolling degree day sum", rolling_degree_day_sum
    
    if DEBUG: print "forecast_degree_day_sum=", forecast_degree_day_sum
    return forecast_degree_day_sum

    
def test_degree_day_sum_wonline():
    degree_day_sums = get_degree_day_sum_wonline('94538', '2013-1-1', 4)
    print "\n ****Degree day sums wonline (yesterday + today + 9 days) *****"
    print degree_day_sums
    print ""


def test_chilling_degree_day_sum_wonline():
    degree_day_sums = get_degree_day_sum_wonline('94538', '2013-08-15', 20, chilling=True)
    print "\n ****Chilling degree day sums wonline (yesterday + today + 9 days) *****"
    print degree_day_sums
    print ""
    
test_degree_day_sum_wonline()
test_chilling_degree_day_sum_wonline()
```

#Making real time predictions
We can now use our model to make predictions. The model currently covers the area around Boston so we will limit our predictions to be around that area. We will make predictions about multiple leafing events such as bud break and most intense fall colors. A number of helper methods are involved in generating the predictions. Following is a run down of the tasks involved:

###Predictions Display Options: Cumulative vs. Discrete
The model provides probabilities of events in a cumulative fashion. This means that once the probabilities of a specific event is given, predictions for the next day would be incremental. Over the course of several days, if the event is gaining strength, the probabilities will keep on increasing until the event happens. At that point, the probabilities will flatten out. This is much easier explained when you see the actual graph which displays the probabilities (in the validation section).

Another way to display the probabilities is to keep a rolling counter by subtracting the prvious day's probability from the current day and so on. In this type of graph you would be able to see the probabilities go up and down as they hit the max probability of the event and then flatten out.

###A brief description of methods

####- get_predictions (lat, lng, end_date=None, cumulative=False): 
Returns the predictions given a latitude and longitude. If end date is None, gets data uptil today and then gets the next 10 days forecast for the "true" future. If a past end date is given (e.g. June 2, 2013), then all historic data until June 2 is fetched from history. The method will then fetch the next 10 days data after June 2 as forecast data. We use this mechanism so that we can get forecast data transparently for "true" forecast dates as well as past forecast dates. Using past forecasts allows us to check the validity of the model. Cumulative flag dictates whether to return cumulative or discrete representation of the probabilities.


####- get_predictions_by_zip(zip_code, end_date=None, cumulative=False):
Same thing as the previous method, uses zip code instead of latitude and longitude.


####- get_gdd_start_date(end_date=None):
Spring events are calclulated by using growing/heating degree days. We start aggregating gdd temperature starting with January 1. If the end date is none, the aggregation is carried out from January 1 to the current day. If the end date is given (for example June 15) we will return gdd temperature aggreation from January 1 - June 15.


####- get_chilling_start_date(end_date=None):
Fall events are calculated by aggregated chilling days below a certain threshold. The start date for chilling days is August 15. The end date is used to control how many days to carry out the aggregation. If end date is None and the api is called on Dec. 15, then we will aggregate the chilling days from August 15, 2013 to Dec. 15, 2013. 


####- get_event_type_predictions(event_type, gdd_arr, cumulative=False):
Invokes the method to get the probability for different events. The models are packaged in multiple files (named as their respective events) and are invoked appropriately by the application. The argument, gdd_arr, is used to provide either heating or chilling days accumulated temperatures.


####- get_zero_probabilities(probs_in_array):
When we are returning the probabilities for spring events, the fall events have  probability of zero and vice versa. This utility method allows us to tack on zero probabilities for those superfluous entries. This is mostly useful for web client display. For example, if the user asked for probabilities when the end date is October 11, we will just set the probabilities of spring events as zero, calculate the accumulated chilling degrees and return probabilities of fall events: fall colors, fall intense colors and leaf drop dates.


####- convert_from_cumulative_to_discrete(event_predictions):
This method does a rolling subtraciton to convert cumulative to discrete predictions. Each representation seem to have its own advantages. The validation sample below will use the cumulative distribution as it seems to convey a pretty good visual representation of the progression of events. 

```python
"""
Finally we make our predictions for different leaf events.

Following are the selected events for the leaves:

- Bud Break
- Leaf Maturity (Greenest)
- Beginning of fall colors
- Most Intense Fall Colors
- End of leaf drop
"""
def convert_from_cumulative_to_discrete(event_predictions):
    discrete_predictions = []
    for idx, p in enumerate(event_predictions):
        if idx == len(event_predictions) - 1:
            break
        discrete_p = event_predictions[idx + 1] - event_predictions[idx]
        discrete_p = "%.2f" % discrete_p
        discrete_predictions.append(discrete_p)
    return discrete_predictions


        
def is_end_date_greater(start_date, end_date):
    start_date_date = datetime.datetime.strptime(start_date, "%Y-%m-%d").date()
    end_date_date = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()    
    return end_date_date > start_date_date
        


# After Aug 15, we are only interested in fall events and vice versa
# So we just put zero probabilities for the events we do not consider
def get_zero_probabilities(probs_in_array):
    zero_probs_arr = []
    for i in range(probs_in_array):
        zero_probs_arr.append(0.0)
    return zero_probs_arr



def get_event_type_predictions(event_type, gdd_arr, cumulative=False):
    
    event_predictions = None
    model_file_name = None
    
    if event_type == "start_of_spring": #Bud Break
        model_file_name = "MODIS/start_of_spring_model2007.pkl"
    elif  event_type == "end_of_spring": #Leaf Maturity (Greenest)
        model_file_name = "MODIS/end_of_spring_model2007.pkl"
    elif  event_type == "start_of_fall": #Beginning of fall colors
        model_file_name = "MODIS/start_of_fall_model2007.pkl"    
    elif  event_type == "middle_of_fall": #Most Intense Fall Colors
        model_file_name = "MODIS/middle_of_fall_model2007.pkl"
    elif  event_type == "end_of_fall": #End of leaf drop
        model_file_name = "MODIS/end_of_fall_model2007.pkl"
    
    #feed the values to the model
    model_clone = joblib.load(model_file_name)

    #Get right number of dimesions and shape for non-dim and 1-dim
    gdd_arr = np.ma.atleast_2d(gdd_arr)
    gdd_arr = np.transpose(gdd_arr)
    event_predictions = model_clone.predict_proba(gdd_arr)[:,1]
    
    if DEBUG: print "event_predictions", event_predictions
    discrete_predicitons = convert_from_cumulative_to_discrete(event_predictions)
    if DEBUG: print "discrete_predicitons", discrete_predicitons
    
    if cumulative:
        return event_predictions
    else:
        return discrete_predicitons


# Degree day calculations are from the start of the year
# So if someone looks at March 1, 2009 - we start accumulating from Jan. 1, 2009
def get_gdd_start_date(end_date=None):

    OLDEST_YEAR_SUPPORTED = 2009

    DEF_START_MONTH = "1"
    DEF_START_DAY = "1"
    DEF_START_YEAR = "2013"
    DEF_START_DATE = DEF_START_YEAR + "-" + DEF_START_MONTH + "-" + DEF_START_DAY
    
    if end_date is None:
        return DEF_START_DATE
    
    start_date = None
    try:
        start_year = end_date.split("-")[0]
        
        if int(start_year) < OLDEST_YEAR_SUPPORTED: 
            return DEF_START_DATE
        
        start_date = start_year + "-" + DEF_START_MONTH + "-" + DEF_START_DAY
        # check if this is a real date and end date > start date
        start_date_check = datetime.datetime.strptime(start_date, "%Y-%m-%d").date()
        end_date_check = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()
        if start_date_check > end_date_check:
            return DEF_START_DATE
    except:
        return DEF_START_DATE
    
    return start_date



def get_chilling_start_date(end_date=None):

    OLDEST_YEAR_SUPPORTED = 2009

    DEF_START_MONTH = "8"
    DEF_START_DAY = "15"
    DEF_START_YEAR = "2013"
    DEF_START_DATE = DEF_START_YEAR + "-" + DEF_START_MONTH + "-" + DEF_START_DAY
    
    if end_date is None:
        return DEF_START_DATE
    
    start_date = None
    try:
        start_year = end_date.split("-")[0]
        
        if int(start_year) < OLDEST_YEAR_SUPPORTED: 
            return DEF_START_DATE
        
        start_date = start_year + "-" + DEF_START_MONTH + "-" + DEF_START_DAY
        # check if this is a real date and end date > start date
        start_date_check = datetime.datetime.strptime(start_date, "%Y-%m-%d").date()
        end_date_check = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()
        if start_date_check > end_date_check:
            return DEF_START_DATE
    except:
        return DEF_START_DATE
    
    return start_date
    



def get_predictions_by_zip(zip_code, end_date=None, cumulative=False):
    
    if DEBUG: print "**EXEC: get_predictions_by_zip(zip_code, end_date=None):", \
        zip_code, end_date
        
    lat, lng = get_lat_long_for_zip(zip_code)
    return get_predictions(lat, lng, end_date, cumulative)



def get_predictions (lat, lng, end_date=None, cumulative=False): 
 
    if DEBUG: print "**EXEC: get_predictions (lat, lng, end_date=None):", \
        lat, lng, end_date
    
    event1_predictions, event2_predictions, event3_predictions, event4_predictions, event5_predictions = \
        None, None, None, None, None

    GROWING_BASE_TEMP = 4
    GDD_START_DATE = get_gdd_start_date(end_date)
    if DEBUG: print "GDD_START_DATE", GDD_START_DATE
        
    CDD_START_DATE = get_chilling_start_date(end_date)
    if DEBUG: print "CDD_START_DATE", CDD_START_DATE
    
    json_resp_dict = {}
    json_resp_dict['lat'] = lat
    json_resp_dict['lng'] = lng
    
    zip_code = get_zip_for_lat_long(lat, lng)    
   
    #get growth degree days
    gdds = get_degree_day_sum(zip_code, GDD_START_DATE, GROWING_BASE_TEMP, end_date)
    
    #extract gdds out of the dictionary
    gdd_arr = [gdd.values()[0] for gdd in gdds]
    if is_end_date_greater(CDD_START_DATE, end_date):
        event1_predictions, event2_predictions = get_zero_probabilities(11), get_zero_probabilities(11)
    else:
        event1_predictions = get_event_type_predictions("start_of_spring", gdd_arr, cumulative)
        event2_predictions = get_event_type_predictions("end_of_spring", gdd_arr, cumulative)
    

    #get chilling degree days
    CHILLING_BASE_TEMP = 20
    if not is_end_date_greater(CDD_START_DATE, end_date):
        event3_predictions, event4_predictions, event5_predictions = \
            get_zero_probabilities(11),get_zero_probabilities(11),get_zero_probabilities(11)
    else:
        cdds = get_degree_day_sum(zip_code, CDD_START_DATE, CHILLING_BASE_TEMP, end_date, True) #chilling
        cdd_arr = [cdd.values()[0] for cdd in cdds]
        event3_predictions = get_event_type_predictions("start_of_fall", cdd_arr, cumulative)
        event4_predictions = get_event_type_predictions("middle_of_fall", cdd_arr, cumulative)
        event5_predictions = get_event_type_predictions("end_of_fall", cdd_arr, cumulative)

        
    predictions_arr = []
    for gdd, event_1, event_2, event_3, event_4, event_5 in \
    zip(gdds, event1_predictions,event2_predictions, event3_predictions, event4_predictions, event5_predictions):

        events_container = {}
        events_dict = {}
        events_dict['event_1'] = event_1
        events_dict['event_2'] = event_2
        events_dict['event_3'] = event_3
        events_dict['event_4'] = event_4
        events_dict['event_5'] = event_5
        
        events_container['date'] = gdd.keys()[0]
        events_container['events'] = events_dict
        
        predictions_arr.append(events_container)
        if len(predictions_arr) == 11: 
            predictions_arr = predictions_arr[1:]
        
    json_resp_dict['predictions'] = predictions_arr
    
    return json_resp_dict


## test main methods with different combinations

#get_predictions("42.446396", "-71.459405", "2013-9-15")
get_predictions("42.446396", "-71.459405", "2013-9-15", cumulative = True)
#get_predictions("42.446396", "-71.459405", "2013-5-01")
#get_predictions("42.446396", "-71.459405", "2013-5-01", cumulative = True)
#get_predictions_by_zip("02238", "2013-3-15")
#get_predictions_by_zip("02238", "2013-3-15", cumulative = True)
#get_predictions_by_zip("02238", "2013-10-01")
#get_predictions_by_zip("02238", "2013-10-01", cumulative = True)
##get_predictions_by_zip("02238", "2008-6-15") # error - cannot go before 2009


## testing supporting methods
#get_zero_probabilities(11)
#get_chilling_start_date("2011-9-15")
#get_event_type_predictions("end_of_fall", [200, 201, 202])
```

As the predicted cumulative probabilities, shown in green, increase from zero to one, the probability of a leaf event having happenned increases.  When the probability crosses 0.5, the model is predicting that the event is likley to have happened by then.  The first four phenological events we identified in our subjective assessment are within two weeks of the date of 0.5 cumulative probability, indicating reasonable model performance, given the subjective nature of the validation criteria.  The last event, leaf drop, is somewhat further from the predicted date of 0.5 cumulative probability.  Leaf drop is difficult to predict because the separation of leaves from branches depends on physical factors, such as wind and rain, in addition endogenous biological factors within the tree.  So this event is expected to be more difficult to predict using the degree day model.
